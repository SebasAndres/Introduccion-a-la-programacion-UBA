* Partes de una especificacion:
    * Encabezado.
    * <b>Precondicion (requiere)</b>: Describen todas las condiciones y posibles valores o casuisticas de los parametros de entrada.

## Teorica 2: Logica

#### Formulas, Simbolos y Variables proposicionales

> Formulas: Son los True, False y las variables proposicionales con sus operaciones validas.

> Simbolos: $\rightarrow, \exists, \forall$

> Variables Proposicionales: $p, q, ...$

#### Relacion de fuerza entre formulas.
> Decimos que $A$ es mas fuerte que $B$ cuando $A \rightarrow B$ es tautologia

#### Equivalencias entre formulas:

1. Idempotencia:
* $(p∧p) ↔p$ <br>
* $(p∨p) ↔p$

2. Asociatividad:
* $(p∧q)∧r ↔ p∧(q∧r)$ <br>
* $(p∨q)∨r ↔ p∨(q∨r)$

3. Conmutatividad:
* $(p∧q) ↔(q∧p)$ <br>
* $(p∨q) ↔(q∨p)$

4. Distributividad:
* $p∧(q∨r) ↔ (p∧q)∨(p∧r)$<br>
* $p∨(q∧r) ↔ (p∨q)∧(p∨r)$

5. Reglas de De Morgan:
* $¬(p∧q) ↔ ¬p ∨ ¬q$ <br>
* $¬(p∨q) ↔ ¬p ∧ ¬q$

#### Tipos de formulas:
> <b>Tautologia</b>:  Una formula es una tautologia si siempre toma el valor V para valores definidos de sus variables proposicionales.

> <b>Contradiccion</b>: Una formula es una contradiccion si siempre toma el valor F para valores definidos de sus variables proposicionales.

> <b>Contingencia</b>: Cuando no es ni tautologia ni contradiccion.

## Teorica 3: Especificacion

#### Sobre-especificacion y Subespecificacion

> <b>Sobre-especificacion</b>: Consiste en dar una postcondicion mas restrictiva de la que se necesita, o bien dar una precondicion mas laxa. Limita los posibles algoritmos que resuelven el problema, porque impone mas condiciones para la salida, o amplıa los datos de entrada.

> <b>Sub-especificacion</b>: Consiste en dar una precondicion mas restrictiva de lo realmente necesario, o bien una postcondicion mas debil de la que se necesita. Deja afuera datos de entrada o ignora condiciones necesarias para la salida (permite soluciones no deseadas).

#### Paradigmas de la Programacion

* <b>Paradigma Declarativo</b>: Son lenguajes donde el programador le indicara a la maquina lo que quiere hacer y el resultado que desea, pero no necesariamente el como hacerlo. Describe un conjunto de condiciones, proposiciones, afirmaciones, restricciones, ecuaciones o transformaciones que describen el problema y detallan su solucion.
    * <b>Logico</b>: Los programas estan construıdos unicamente por expresiones logicas (es decir, son Verdaderas o Falsas)
    * <b>Funcional</b>: Esta basado en el modelo matematico de composicion funcional. El resultado de un calculo es la entrada del siguente, y ası sucesivamente hasta que una composicion produce el valor deseado.

* <b>Paradigma Imperativo</b>: Describe la programacion como una secuencia de instrucciones o comandos que cambian el estado de un programa.
    * <b>Estructurado</b>: los programas se dividen en bloques (procedimientos y funciones), que pueden o no comunicarse entre sı. Existen estructuras de control, que dirigen el flujo de ejecucion: IF, GO TO, Ciclos, etc
    * <b>Orientado a Objetos</b>: Se basa en la idea de encapsular estado y comportamiento en objetos. Los objetos son entidades que se comunican entre sı por medio de mensajes

Existen lenguajes multiparadigma:

## Teorica 8: Validacion y Verificacion (V&V)

#### Control de calidad de software

V&V deberıa aplicarse en cada instancia del proceso de desarrollo.

> <b>Validacion</b>: ¿Estamos haciendo el producto correcto? El software deberıa hacer lo que el usuario requiere de el.

> <b>Verificacion</b>: ¿Estamos haciendo el producto correctamente? El software deberıa realizar lo que su especificacion indica.
 
#### Error $\rightarrow$ Defecto $\rightarrow$ Falla

> <b>Error</b>: Equivocacion humana.

> <b>Defecto</b>: Desperfecto en algun componente del sistema (en el texto del programa, una especificacion, un diseño, etc), que origina una o mas fallas.

> <b>Falla</b>: Diferencia entre los resultados esperados y reales.

#### Objetivos V&V:
* Hallar defectos
* Asegurar que respeca especificacion
* Asegurar que satisface necesidades de usuarios

#### Tipos de V&V

* <b>Estatica</b>: trata con el analisis de una representacion estatica del sistema para descubrir problemas.

* <b>Dinamica</b>: trata con ejecutar y observar el comportamiento de un producto (ej. testing)

#### V&V Dinamica por Testing:

> <b>Testing</b>: Es el proceso de ejecutar un producto para:
>> [1] Verificar que satisface los requerimientos (en nuestro caso, la especificacion) <br>
>> [2] Identificar diferencias entre el comportamiento real y el comportamiento esperado. <br>
Es decir, iene como objetivo encontrar defectos en el software. Representa entre el 30 % al 50 % del costo de un software confiable.

#### Niveles de testing:
> <b>Test de Sistema</b>: Test de todo el sistema (por lo general, de aceptacion).

> <b>Test de Integracion</b>: Esta orientado a verificar que las partes de un sistema que funcionan bien aisladamente,
tambien lo hacen en conjunto (testeamos la interaccion entre las partes).

> <b>Test de Unidad</b>: Se realiza sobre una unidad de codigo pequena, claramente definida.

#### Conceptos del testing:
> <b>Test Input (o dato de prueba)</b>: Es una asignaci ́on concreta de valores a los parametros de entrada para ejecutar el programa bajo test.

> <b>Test Case (o caso de prueba)</b>: Es un programa que ejecuta el programa bajo test usando un dato de test, y chequea (automaticamente) si se cumple la condicion de aceptacion sobre la salida del programa bajo test.

> <b>Test Suite</b>: Es un conjunto de casos de Test (o de conjunto de casos de prueba).

#### Limitaciones del Testing:
Al no ser exhaustivo, el testing NO puede probar (demostrar) que el software funciona correctamente.

Una de las mayores dificultades es encontrar un conjunto de tests
adecuado:
* Suficientemente grande para abarcar el dominio y maximizar la probabilidad de encontrar errores.
* Suficientemente peque ̃no para poder ejecutar el proceso con cada elemento del conjunto y minimizar el costo del testing.

#### Criterios para seleccionar test input:
* <b>Caja Blanca</b>: los casos de test se generan analizando la implementacion para determinar los casos de test.
* <b>Caja Negra</b>: los casos de test se generan analizando la especificacion sin considerar la implementacion.


## Teorica 9: Caja Negra

#### Metodo de Particion de Categorias:

Es una tecnica para generar casos de prueba de una manera metodica.

* Paso 1: Listar todos los programas a testear.
* Paso 2: Elegir un programa de la lista,
    * Paso 3: Identificar los factores (sus parametros y relaciones)
    * Paso 4: Determinar las categorias (caracteristicas relevantes) de cada factor.
    * Paso 5: Determinar las choices por cada categoria de cada factor.
    * Paso 6: Clasificar las choices: errores unicos, restricciones, etc.
    * Paso 7: Armar casos de prueba combinando distintas elecciones para cada categoria y detallando el resultado esperado de c/u.
    * Paso 8: Volver al paso 2 hasta testear todos los programas.


## Teorica 10: Programacion Imperativa

#### Tipos de pasaje de parametros:
* <b>IN</b>: Al salir de la funcion o procedimiento, la variable pasada como parametro continuara teniendo su valor original.

* <b>OUT</b>: Al salir de la funcion o procedimiento, la variable pasada como parametro tendra un nuevo valor asignado dentro de dicha funcion o procedimiento. Su valor inicial no importa ni deberıa ser leıdo dentro de la funci ́on o procedimiento en cuestion.

* <b>INOUT</b>: Al salir de la funcion o procedimiento, la variable pasada como parametro tendra un nuevo valor asignado dentro de dicha funci ́on o procedimiento. Su valor inicial SI importa dentro de la funcion o procedimiento en cuestion.

#### Pasajes de parametros en Python:
* <b>REFERENCIA</b>: Tipos compuestos y estructuras (listas, etc): se pasan por referencia.
* <b>VALOR</b>: Tipo primitivos (int, char, strings, etc) se pasan por valor.

#### Funcion != Procedimiento
* Funcion $\rightarrow$ Devuelve un valor
* Procedimiento $\rightarrow$ No devuelve nada

## Teorica 14: Caja Blanca

#### Saber diagramar Control Flow Graph (CFG)

#### Usos de Caja Blanca:
* Complemento al Test de Caja Negra: permite encontrar mas casos o definir casos mas especificos
* Como criterio de adecuacion del Test de Caja Negra: brinda herramientas que nos ayudar a determinar cuan bueno o confiable resultaron ser los test suites definidos.

#### Criterios de Adecuacion Estructurales por CFG:

* [a] Cubrimiento de Sentencias (nodos): 
    * Definicion: Cada bloque/sentencia/nodo debe ser ejecutado al menos 1 vez por algun test case.
    * Cobertura: $\frac{nodosEjercitados}{totalNodos}$
    
    <br>
    
* [b] Cubrimiento de Arcos (flechas):
    * Definicion: Cada arco/flecha que une a los nodos debe ser ejecutada al menos 1 vez por algun test case.
    * Observacion: El cubrimiento de todos los arcos implica cubrir todos los nodos.
    * Cobertura: $\frac{arcosEjercitados}{totalArcos}$
    
    <br>

* [c] Cubrimiento de Decisiones (branches):
    * Definicion: Todas las deciciones (los arcos True y False en el CFG) deben ser ejecutados por algun test suite.
    * Cobertura: $\frac{decisionesEjercitadas}{totalDeciciones}$
    * Necesitan cobertura 100%
    
    <br>

* [d] Cubrimiento de Condiciones Basicas:
    * Definicion: Una condicion básica es una formula atómica (indivisible) que compone una division. 
    
    * Cobertura: $\frac{CantidadVariablesEvaluadasEnCadaCondicion}{2*CantidadCondicionesBasicas}$
    * Necesitan cobertura 100%

    * Ejemplo: Sea el condicional "(($x=1 || y=2$) ^ len > 0)", las condiciones basicas son...
        * $x=1$
        * $y=2$
        * $len > 0$
    
    <br>

* [e] Cubrimiento de Caminos:
    * Definicion: Cada camino en el CFG debe ser transitado por al menos un test case.
    * Cobertura: $\frac{caminosTransitados}{cantidadCaminos}$
    
    <br>
